**Bar** **plot**

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(13, 8))

sns.barplot(data=df, x='speeding', y='total', ax=axes[0, 0])
axes[0, 0].set_title('Speeding vs Total')
sns.barplot(data=df, x='alcohol', y='total', ax=axes[0, 1])
axes[0, 1].set_title('Alcohol vs Total')
sns.barplot(data=df, x='not_distracted', y='total', ax=axes[1, 0])
axes[1, 0].set_title('Not Distracted vs Total')
sns.barplot(data=df, x='no_previous', y='total', ax=axes[1, 1])
axes[1, 1].set_title('No Previous vs Total')
#Inference:above 4 graphs relate accidents more alcohol intake causes more
#accidents,number of drivers who have records of previous accidents were
#increasing.
```

**scatter** **plot**

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(13, 8))
plt.subplots_adjust(hspace=0.4,wspace=0.4)
sns.scatterplot(x="alcohol",y="not_distracted",data=df,color="red",ax=axes[0,0])
sns.scatterplot(x="ins_losses",y="ins_premium",data=df,color="green" ,ax=axes[0,1])
sns.scatterplot(x='total',y='speeding',data=df ,ax=axes[0,2])
sns.scatterplot(x='total',y='alcohol',data=df,color="pink" ,ax=axes[1,0])
sns.scatterplot(x='total',y='not_distracted',data=df ,color="violet",ax=axes[1,1])
sns.scatterplot(x='total',y='no_previous',data=df ,color="skyblue",ax=axes[1,2])
sns.scatterplot(x='alcohol',y='no_previous',data=df,color="lightgreen",ax=axes[2,0])
sns.scatterplot(x='alcohol',y='speeding',data=df,color="orange",ax=axes[2,1])
sns.scatterplot(x='no_previous',y='speeding',data=df,color="black",ax=axes[2,2])
```

inference-A scatter plot in Matplotlib is a type of data visualization that displays individual data points as dots on a two-dimensional plane, with one dot for each data point. It is used to visualize the relationship or distribution of two continuous variables. Each dot on the scatter plot represents a single data point, and the position of the dot is determined by the values of the two variables being compared.

**count plot**

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(13, 8))
plt.subplots_adjust(hspace=0.4,wspace=0.4)
sns.countplot(x="speeding",data=df,ax=axes[0,0])
sns.countplot(x="alcohol",data=df,ax=axes[0,1])
sns.countplot(x="total",data=df,ax=axes[1,0])
sns.countplot(x="not_distracted",data=df,ax=axes[1,1])
```

**inference**-This graph depicts the distribution of car accidents involving and not involving speeding. It enables you to estimate the frequency of speeding-related crashes.

**line plot**

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(13, 8))
plt.subplots_adjust(hspace=0.4,wspace=0.4)
sns.lineplot(x='total',y='speeding',data=df,color="red",ax=axes[0,0])
sns.lineplot(x='total',y='alcohol',data=df,color="blue",ax=axes[0,1])
sns.lineplot(x='alcohol',y='speeding',data=df,color="lightgreen",ax=axes[1,0])
sns.lineplot(x="ins_premium",y="ins_losses",data=df,color="cyan")
```

inference -

plot-1-On average speeding id directly proportional to car crashes

plot-2-as the alcohol consumption is increasing the total deaths are also increasing which indicates more crashes are due to alcohol

plot-3-inference is that when alcohol increases speed also increases

plot-4-Inference: On average, Premium increases ideally losses decreases

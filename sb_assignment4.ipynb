```python
x = df.drop(columns=['Attrition'])
y = df['Attrition']
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x["BusinessTravel"]=le.fit_transform(x["BusinessTravel"])
x["Department"]=le.fit_transform(x["Department"])
x["EducationField"]=le.fit_transform(x["EducationField"])
x["Gender"]=le.fit_transform(x["Gender"])
x["JobRole"]=le.fit_transform(x["JobRole"])
x["MaritalStatus"]=le.fit_transform(x["MaritalStatus"])
x["Over18"]=le.fit_transform(x["Over18"])
x["OverTime"]=le.fit_transform(x["OverTime"])
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.256,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#1 logistic regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
predlr=model.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
accuracy_score(y_test,predlr)
```

```python
confusion_matrix(y_test,predlr)
```

```python
pd.crosstab(y_test,predlr)
```

```python
print(classification_report(y_test,predlr))
```

```python
#2 desicion tree classifier
from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy="auto", random_state=42)
x1, y1 = smote.fit_resample(x, y)
x_train1,x_test1,y_train1,y_test1=train_test_split(x1,y1,test_size=0.189,random_state=0)
x_train1=sc.fit_transform(x_train1)
x_test1=sc.fit_transform(x_test1)
model2=DecisionTreeClassifier()
model2.fit(x_train1, y_train1)
```

```python
preddt = model2.predict(x_test1)
```

```python
accuracy_score(y_test1,preddt)
```

```python
print(confusion_matrix(y_test1,preddt))
print(pd.crosstab(y_test1,preddt))
print(classification_report(y_test1,preddt))
```

```python
#3 random forest classifier
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy="auto", random_state=42)
x3, y3 = smote.fit_resample(x, y)
x_train3,x_test3,y_train3,y_test3=train_test_split(x3,y3,test_size=0.25009,random_state=0)
x_train3=sc.fit_transform(x_train3)
x_test3=sc.fit_transform(x_test3)
model3= RandomForestClassifier(n_estimators=100, random_state=42)
model3.fit(x_train3, y_train3)
```

```python
predrf=model3.predict(x_test3)
y_pred_binary = [1 if score > 0.5 else 0 for score in predrf]
accuracy_score(y_test3, y_pred_binary)
```

```python
print(confusion_matrix(y_test3,y_pred_binary))
print(pd.crosstab(y_test3,y_pred_binary))
print(classification_report(y_test3,y_pred_binary))
```
